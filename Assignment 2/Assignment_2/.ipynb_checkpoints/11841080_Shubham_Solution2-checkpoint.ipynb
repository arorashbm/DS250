***
This is a multi-step process:
    1. Load a dataset from `geopandas` library which contains dimensions and shapes requires to create world
    map. This is done in cell below, and stored in dataframe `world`
    
    2. Next we want a column with 3-letter country code, to merge our data with dataframe `world`, which is
    done by function `makeDataForMap` (described below).
    
    3. Next merge our dataframe with `world` on 3-letter country code.

```python
# load dataset 'naturalearth_lowres' from geopandas
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
```

***
<div class="alert alert-block alert-info">
<b>Function Description:&nbsp;</b> This function converts 2-letter country codes given in dataset to 3-letter code using module <strong>pycountry</strong>. <br/>
    Then it merges current dataframe with <strong>world</strong> on 3-letter country code, and return that dataframe. <br/>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<strong>Input:&nbsp;</strong>Name of feature to used in map <br/>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<strong>Output:&nbsp;</strong>Pandas DataFrame
</div>

```python
def makeDataForMap(col):
    '''
        This function converts 2-letter country codes given in dataset to 3-letter code using module pycountry.
        Then it merges current dataframe with world on 3-letter country code, and return that dataframe.
        
        Input: Name of feature to used in map
        Output: Pandas DataFrame
    '''
    
    Country_Code_2 = data["Country Code"]          # 2-letter country code
    data_ret = []                                  # list to be used in creating dataframe
    col_data = data[col]                           # data of column passed to function
    i = 0

    for code in Country_Code_2:
        # if 2-letter code is valid append 3-letter and col value for that country to list
        try:
            country = pycountry.countries.get(alpha_2 = code)
            data_ret.append([country.alpha_3, col_data[i]])
        except:
            data_ret.append([None, col_data[i]])
        i += 1
        
    # make pandas dataframe with 3-letter code and value of col for that country
    data_for_map = pd.DataFrame(data_ret, columns=["iso_a3", col])
    
    # merge dataframes based on 3-letter code
    data_for_map = pd.merge(world, data_for_map, on = "iso_a3")
    
    # return resulting dataframe
    return data_for_map
```

```python
# feature name used in map
col_name = "Unemployment rate"

# make required dataframe by calling makeDataForMap
data_for_map = makeDataForMap(col_name)

# plot data and give it title
data_for_map.plot(column = col_name, figsize=(40, 40), scheme = "quantiles", legend = True)
plt.title("Unemployment Rate for different countries", size = 40)
plt.show()
```

<font size="5">[Computing Correlation Coefficient](#computing-correlation-coefficient)</font>

***
<div class="alert alert-block alert-info">
<b>Function Description:&nbsp;</b> This function make dataframe from correlation matrix, with given format <span style="background:#FFFF00;color:red">Variable1, Variable2,Correlation_Coefficient</span> <br/>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<strong>Input:&nbsp;</strong>Correlation Matrix <br/>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<strong>Output:&nbsp;</strong>1. Pandas Dataframe, 2. List with top 100 pearson coefficient
</div>

```python
def GetMax20(data):
    '''
        This function make dataframe from correlation matrix, with given format 
        Variable1, Variable2,Correlation_Coefficient
        
        Input: Correlation Matrix
        Output: 1. Pandas Dataframe, 2. List with top 100 pearson coefficient
    '''
    n = len(data)         # length od data
    cols = data.columns   # columns od data
    values = []           # storing variable1, variable2, correlation coefficient in list
    for i in range(n):
        row = data.iloc[i]
        for key in range(i+1, len(row)):
            if i != key:
                # append to variable1, variable2, correlation coefficient list
                values.append([cols[i], cols[key], row[key]])
    values.sort(key = lambda value: value[2], reverse = True)   # sort values to store top 20 and top 100
    values_100 = []       # store top 100 correlation coefficient
    count = 0
    l = 0
    r = len(values)-1
    while count < 100:
        # both positive and negative can be strong, therefore taking absolute value
        if abs(values[l][2]) < abs(values[r][2]):
            values_100.append(values[r])
            r -= 1
        else:
            values_100.append(values[l])
            l += 1
        count += 1
    # storing top 20 in dataframe
    df = pd.DataFrame(values_100[:20], columns = ["Variable1", "Variable2", "Correlation_Coefficient"])
    return df, values_100   # returning dataframe and top 100 correlation coefficient
```

***
Constructing Correlation matrix with `pearson` and `spearman` methods

```python
# computing correlation with corr() function of pandas
pearson = data.corr(method="pearson")
spearman = data.corr(method="spearman")
```

***
Calling function `GetMax20` to get dataframe with top 20 **pearson** and **spearman** coefficient dataframes
, along with lists of top 100 **pearson** and **spearman** coefficients lists, to be used later

```python
# computing top 20 correlation that correlate strongest and top 100 correlation list
pearson_20_df, pearson_100 = GetMax20(pearson)
spearman_20_df, spearman_100 = GetMax20(spearman)
```

As observed from top 20 **pearson** and **spearman**, correlation coefficient of **spearman** is always less than or equal to **pearson**, because **pearson** does not take into account the effect of outliers, but **spearman** does take *outliers* into account while calculating it, hence the difference between **pearson** and **spearman** coeffcient.

<font size="5">[Performing Permutations Test](#performing-permutations-test)</font>

***
<div class="alert alert-block alert-info">
<b>Function Description:&nbsp;</b> This function performs permutations test on dataframe passed to it in form <span style="background:#FFFF00;color:red">Variable1, Variable2, Correlation_Coefficient</span> and return dataframe of form <span style="background:#FFFF00;color:red">Variable1, Variable2, P Value, Significance, Fraction of permutations with greater than given correlation, Fraction of Permutations produce at least this high a correlation</span> <br/>
    A total of 10000 random permutations are generated in the test from 197! permutations, that's why result is very small (tends to 0) <br/>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<strong>Input:&nbsp;</strong>Pandas DataFrame <br/>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<strong>Output:&nbsp;</strong>Pandas DataFrame with computed values
</div>

```python
def CalcPValue(pearson_100):
    '''
        This function performs permutations test on dataframe passed to it in form 
        Variable1, Variable2, Correlation_Coefficient and return dataframe of form 
        Variable1, Variable2, P Value, Significance, Fraction of permutations with greater 
        than given correlation, Fraction of Permutations produce at least this high a correlation
        
        Input: Pandas DataFrame
        Output: Pandas DataFrame with computed values
    '''
    
    df_list = []     # list to make dataframe
    count = 0
    for p_100 in pearson_100:
        count += 1
        var1 = p_100[0]
        var2 = p_100[1]
        # storing columns of variable1, variable2 to col_var1, col_var2
        col_var1 = data[var1]
        col_var2 = data[var2]
        
        # computing p value from function permutation_test with only chossing 10000 permutations randomly
        # out of 197! permutations
        p_value = permutation_test(col_var1, col_var2, method='approximate', num_rounds=10000, func=lambda x, y: np.corrcoef(x, y)[1][0], seed=0)

        # appending df_list to be used in making dataframe in next step
        df_list.append([var1, var2, p_value, p_value*100, max(0, p_value-0.0001), 0])
    
    # make dataframe from df_list and return it
    df = pd.DataFrame(df_list, columns=["Variable1", "Variable2", "P Value", "Significance", "Fraction of permutations with greater than given correlation", "Fraction of Permutations produce at least this high a correlation"])
    return df
```

Call `CalcPValue` with top 100 list by `pearson's method`

```python
# calculate p-value, significance, fraction with correlation greater than given
dfPValue = CalcPValue(pearson_100)
```

First, on list which which appear to be normal:

|---------------------------------|----------|
|       Population growth rate    |  0.2977  |

```python
for col in appear_normal:
    # function to perform kstest, "norm" means check for normal distribution
    
    # passing mean and std of our distribution to test against normal
    s, p = kstest(data[col].dropna(), "norm", args=(data[col].mean(), data[col].std()))
    label_str = "p-value: " + str(p)         # string for legend of plot
    sns.displot(normalised_data[col], kind="kde", label = label_str)  # draw distribution plot
    
    # limits on x-y axis of plots
    plt.xlim(-2.5, 3)
    plt.ylim(0,1.3)
    plt.legend()
    plt.show()
```

|                 Roadways                |1.2050e-27|

```python
for col in not_appear_normal:
    # function to perform kstest, "norm" means check for normal distribution
    
    # passing mean and std of our distribution to test against normal
    s, p = kstest(data[col].dropna(), "norm", args=(data[col].mean(), data[col].std()))
    label_str = "p-value: " + str(p)       # string for legend of plot
    sns.displot(normalised_data[col], kind="kde", label = label_str)   # draw distribution plot
    
    # limits on x-y axis of plots
    plt.xlim(-3.5, 4)
    plt.ylim(0,1.2)
    plt.legend()
    plt.show()
    
    print("Distribution looks similar to:")
    
    # using fitter function to find all distributions which are close
    f = Fitter(data[col], distributions='common')
    f.fit()
    f.summary()
```

***
Using Feature ranking with recursive feature elimination, to rank features in linear regression model and selecting those with rank 1 as filter to reduce number of features to be used in next step.

***
Output of this cell prints the rank given to feature, and taking only features (32 features) with rank 1 in the next step

```python
estimator = LinearRegression()       # using linear regression model in RFE to select features
selector = RFE(estimator, step=1, n_features_to_select=32)   # total 32 features are selected as a result
selector = selector.fit(X, y)        # this will fit the model and based on this give rank to each feature
selector.ranking_                    # print ranking of each of 32 features
```

***
The cell below make a list of features to be used in next step

`selected_features` is the name of list of such features

```python
selected_features = []      # list to insert 32 selected features
support = selector.support_ # boolean list with true for features of rank 1, otherwise false

# loop over support list to append list selected_features with column name
for i in range(len(support)):
    if support[i]:
        selected_features.append(cols_reg[i+7])
```

## Training regression model:

***
The model is trained on linear, polynomial (degree = 2), and elastic regression model, and graph is plotted for each to check which model is fit best, which is overfitting etc.

***
<div class="alert alert-block alert-info">
<b>Function Description:&nbsp;</b> This function performs Linear regression by training model on training data (~80% rows) and then result is predicted on test data (~20% rows).<br/>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<strong>Input:&nbsp;</strong>1. Training data, features and labels 2. Test data, features and labels<br/>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<strong>Output:&nbsp;</strong>Training error and test error</div>

```python
def linear_regression(X_train, X_test, y_train, y_test):
    '''
        This function performs Linear regression by training model on training data
        (~80% rows) and then result is predicted on test data (~20% rows)
    '''
    
    lin_reg = LinearRegression()                       # declaring linear model
    lin_reg.fit(X_train, y_train)                      # fit training data
    train_predict = lin_reg.predict(X_train)           # predict for training set
    test_predict = lin_reg.predict(X_test)             # predict for test set

    # compute test and train error
    train_error = mean_squared_error(y_train, train_predict)
    test_error = mean_squared_error(y_test, test_predict)
    
    # return errors, linear model and predictions
    return train_error, test_error, lin_reg, train_predict, test_predict
```

***
<div class="alert alert-block alert-info">
<b>Function Description:&nbsp;</b> This function performs Ploynomial regression (degree = 2) by training model on training data (~80% rows) and then result is predicted on test data (~20% rows)<br/>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<strong>Input:&nbsp;</strong>1. Training data, features and labels 2. Test data, features and labels<br/>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<strong>Output:&nbsp;</strong>Training error and test error</div>

```python
def polynomial_regression(X_train, X_test, y_train, y_test):
    '''
        This function performs Ploynomial regression (degree = 2) by training model on training data 
        (~80% rows) and then result is predicted on test data (~20% rows).
        
        Input: 1. Training data, features and labels 2. Test data, features and labels
        Output: Training error and test error
    '''
    
    poly_reg = PolynomialFeatures(degree=2)                   # declaring polynomial model
    X_poly_train = poly_reg.fit_transform(X_train)            # convert train set to polynomial train set
    X_poly_test = poly_reg.fit_transform(X_test)              # convert test set to polynomial test set
    poly_reg.fit(X_poly_train, y_train)                       # fit polynomial model
    
    # train linear model with polynomial features set
    lin_reg_poly = LinearRegression()                         # declaring linear model       
    lin_reg_poly.fit(X_poly_train, y_train)                   # fit training data
    train_predict = lin_reg_poly.predict(X_poly_train)        # predict for training set
    test_predict = lin_reg_poly.predict(X_poly_test)          # predict for test set
        
    # compute test and train error
    train_error = mean_squared_error(y_train, train_predict)
    test_error = mean_squared_error(y_test, test_predict)
    
    # return errors and prections
    return train_error, test_error, train_predict, test_predict
```

***
<div class="alert alert-block alert-info">
<b>Function Description:&nbsp;</b> This function performs Elastic regression by training model on training data (~80% rows) and then result is predicted on test data (~20% rows)<br/>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<strong>Input:&nbsp;</strong>1. Training data, features and labels 2. Test data, features and labels<br/>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<strong>Output:&nbsp;</strong>Training error, test error and trained elastic model</div>

```python
def elastic_net(X_train, X_test, y_train, y_test):
    '''
        This function performs Elastic regression by training model on training data (~80% rows) and then 
        result is predicted on test data (~20% rows)
        
        Input: 1. Training data, features and labels 2. Test data, features and labels
        Output: Training error, test error and trained elastic model
    '''
    
    elastic_reg = ElasticNet(random_state=0)                 # declaring elastic net model
    elastic_reg.fit(X_train, y_train)                        # fit training data
    train_predict = elastic_reg.predict(X_train)             # predict for training set
    test_predict = elastic_reg.predict(X_test)               # predict for test set
        
    # compute test and train error
    train_error = mean_squared_error(y_train, train_predict)
    test_error = mean_squared_error(y_test, test_predict)
        
    # return errors, elastic net model and predictions
    return train_error, test_error, elastic_reg, train_predict, test_predict
```

***
The cell below calls each regression method with both training and test set as arguments, make a list of root mean square error on training and test set, for each method

To improve accuracy of our model, this next process is looped 100 times to improve accuracy

```python
# initializing lists to store errors of each model
linear_error = []    # list a lists of size 2 for train and test error respectively for linear model
poly_error = []      # list a lists of size 2 for train and test error respectively for polynomial model
elastic_error = []   # list a lists of size 2 for train and test error respectively for elastic net model

# variables to store r2_score and max_r2_score for all 3 models
r2_score_linear = 0
r2_score_poly = 0
r2_score_elastic = 0
max_r2_score_linear = -2
max_r2_score_poly = -2
max_r2_score_elastic = -2

# store list of actual and predicted value at max_r2_score foe all 3 models
train_predict_linear = []
test_predict_linear = []
train_predict_poly = []
test_predict_poly = []
train_predict_elastic = []
test_predict_elastic = []
train_linear = []
test_linear = []
train_poly = []
test_poly = []
train_elastic = []
test_elastic = []

# loop 100 times to improve accuracy
for _ in range(100):
    #split data into trainand test
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 9)
    
    # call linear regression
    train_error, test_error, linear_reg, train_predict, test_predict = linear_regression(X_train, X_test, y_train, y_test)
    linear_error.append([train_error, test_error])
    
    # calculate r2_score
    r2_score_linear += r2_score(test_predict, y_test)
    
    # store data of actual and predict for maximun r2_score
    if r2_score_linear > max_r2_score_linear:
        train_predict_linear = train_predict
        test_predict_linear = test_predict
        train_linear = y_train.tolist()
        test_linear = y_test.tolist()

    # call polynomial_regression for polynomial model
    train_error, test_error, train_predict, test_predict = polynomial_regression(X_train, X_test, y_train, y_test)
    poly_error.append([train_error, test_error])
    
    # calculate r2_score
    r2_score_poly += r2_score(test_predict, y_test)
    
    # store data of actual and predict for maximun r2_score
    if r2_score_poly > max_r2_score_poly:
        train_predict_poly = train_predict
        test_predict_poly = test_predict
        train_poly = y_train.tolist()
        test_poly = y_test.tolist()

    # call elastic_net for elastic net model
    train_error, test_error, elastic_reg, train_predict, test_predict = elastic_net(X_train, X_test, y_train, y_test)
    elastic_error.append([train_error, test_error])
    
    # calculate r2_score
    r2_score_elastic += r2_score(test_predict, y_test)
    
    # store data of actual and predict for maximun r2_score
    if r2_score_elastic > max_r2_score_elastic:
        train_predict_elastic = train_predict
        test_predict_elastic = test_predict
        train_elastic = y_train.tolist()
        test_elastic = y_test.tolist()
```

***
<div class="alert alert-block alert-info">
<b>Function Description:&nbsp;</b> This function given a list of errors generated in previous cell, and compute mean error <br/>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<strong>Input:&nbsp;</strong>List of errors<br/>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<strong>Output:&nbsp;</strong>Mean training and test error</div>

```python
def calcError(error_list):
    '''
        This function given a list of errors generated in previous cell, and compute mean error
        
        Input: List of errors
        Output: Mean training and test error
    '''
    
    train_error = 0
    test_error = 0
    
    # compute mean train/test error
    for error in error_list:
        train_error += error[0]
        test_error += error[1]
        
    # return mean train/test error
    return np.sqrt(train_error/(100)), np.sqrt(test_error/(100))
```

```python
def ActualPredictedBar(actual, predict, model):
    '''
        This function plots a bar graph for actual and predicted value.
        
        Input: List of predicted values
        Output: DataFrame of Actual and Predicted Value
    '''
    
    list_actual_predicted = []            # list to make dataframe of actual and predicted value
    
    # append actual/predicted to list created above
    for i in range(len(actual)):
        list_actual_predicted.append([actual[i], predict[i]])
        
    # make df for the same
    df_actual_predicted = pd.DataFrame(list_actual_predicted, columns=["Actual Test", "Predicted Test"])
    
    # plot bar graph of actual.predicted value
    df_actual_predicted.plot(kind="bar", figsize=(15, 10), color=["blue", "lightblue"])
    plt.xlabel("ith Country")
    plt.ylabel("GDP - per capita (PPP)")
    # if else to choose title for plot based on which model is used
    if model == "elastic":
        plt.title("Actual vs Predicted Value for Elastic Net")
    else:
        plt.title("Actual vs Predicted Value for Linear Model")
    plt.show()
    return df_actual_predicted
```

Make a dataframe with `country name 1, country name 2, distance`, where distance is the eucledian distance between two countries based on features used to cluster them.

```python
list_countries_distance = []        # list to make dataframe

# loop over each country with every country is looping over every other country (except itself) 
# and calculating euclidean distance
for i in range(len(data_growth)):
    for j in range(len(data_growth)):
        # not calculating distance with itself
        if i == j:
            continue
            
        # calculate distance
        dist = scipy.spatial.distance.euclidean(data_growth.iloc[i].values, data_growth.iloc[j].values)
        # append countries name and distance to list
        list_countries_distance.append([countries[i], countries[j], dist])

# make dataframe from list 'list_countries_distance'
data_growth_countries_score = pd.DataFrame(list_countries_distance, columns=["Country Name 1", "Country Name 2", "Distance"])
```

### 2. Dividing countries based on their score in terms of development in telecom industry

***
Columns used to define score

```python
telecom_data_list = ["Telephones - fixed lines", "Telephones - mobile cellular", "Internet users", "Broadband - fixed subscriptions"]
```

***
Calculate score and create a dataframe with country and its score

Score is calculated as sum of score of each in `telecom_data_list` relative to `population`, i.e.

$$score = \sum_{i=1}^{4}{\frac{telecomDataList[i]}{population}}$$

```python
score_country_telecom = []          # list to make dataframe

# loop over each country and calculate score for that country
for i in range(len(data)):
    pop = data.iloc[i]["Population"]
    score = 0
    for tel in telecom_data_list:
        score += (data.iloc[i][tel]/pop)
        
    # make list of lists with country name and score to make dataframe
    score_country_telecom.append([data.iloc[i]["Country Name"], score])
    
    # make pandas dataframe
    df_score_country_telecom = pd.DataFrame(score_country_telecom, columns = ["Country", "Score"])
```

***
Most similar and different country for all countries in the world

```python
country_similar_different = []   # list to store values and then make dataframe from this list

# calculate similar and different country for each country
for i in range(len(score_country_telecom)):
    
    # for first country similar country is just below it and different is last one
    if(i == 0):
        country_similar_different.append([score_country_telecom[i][0], score_country_telecom[i+1][0], score_country_telecom[len(score_country_telecom)-1][0]])
        
    # for last country similar is just above it and different is first one
    elif(i == len(score_country_telecom)-1):
        country_similar_different.append([score_country_telecom[i][0], score_country_telecom[i-1][0], score_country_telecom[0][0]])
    
    # for any other i, for country i, similar is (i+1), if abs(score[i+1]-score[i]) < abs(score[i-1]-score[i])
    # otherwise country i is similar to country (i-1)
    # for country i, country 0 is different if abs(score[0]-score[i]) > abs(score[n-1]-score[i])
    # otherwise country (n-1) is different for country i
    else:
        # similar
        with_i_m_1 = abs(score_country_telecom[i][1]-score_country_telecom[i-1][1])
        with_i_a_1 = abs(score_country_telecom[i][1]-score_country_telecom[i+1][1])
        similar = score_country_telecom[i-1][0] if with_i_m_1 < with_i_a_1 else score_country_telecom[i+1][0]
        
        # different
        with_0 = abs(score_country_telecom[i][1]-score_country_telecom[0][1])
        with_last = abs(score_country_telecom[i][1]-score_country_telecom[len(score_country_telecom)-1][1])
        different = score_country_telecom[0][0] if with_0 > with_last else score_country_telecom[len(score_country_telecom)-1][0]
        
        # append to list
        country_similar_different.append([score_country_telecom[i][0], similar, different])
        
# make dataframe from list 'country_similar_different'
df_country_similar_different = pd.DataFrame(country_similar_different, columns=["Country Name", "Most Similar Country", "Most Different Country"])
```
